```python
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils.np_utils import to_categorical
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler 
import pandas as pd 
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import time
import sys
%matplotlib inline
import random
import copy
import logging
from tqdm import tqdm
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils.np_utils import to_categorical
from keras.callbacks import EarlyStopping
import keras.backend as K
import tensorflow as tf
from tsfresh import extract_features
import tsfresh.feature_extraction.feature_calculators as features
from pandas import DataFrame
import FFNN_Function
import LSTM_Function

# Setup logging.
logger = logging.getLogger()
fhandler = logging.FileHandler(filename='meta-learning.log', mode='a')
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
fhandler.setFormatter(formatter)
logger.addHandler(fhandler)
logger.setLevel(logging.DEBUG)

Serie_names = ['Serie 1','Serie 2','Serie 3','Serie 4','Serie 5','Serie 6','Serie 7','Serie 8','Serie 9','Serie 10','Serie 11','Serie 12']
tree_examples = DataFrame (columns  = ['FFNN','ARIMA','LSTM','length','kurtosis','energy','variance'], index = Serie_names)

```

```python
Series = pd.read_excel("NN3_FINAL_DATASET_WITH_TEST_DATA.xlsx", sheet_name="Hoja 1")
Series
```

```python
steps_ahead = 1
test_size = 14

logging.info("Empezando proceso de llenado de tabla")

for i in Serie_names:
    logging.info("Empieza la serie %s ", i)
    Serie = Series[i].astype(float)
    Serie = Serie.dropna()
    Serie = Serie.values
    Serie = Serie.reshape(Serie.shape[0],-1)
    scaler = MinMaxScaler(feature_range = (0.1, 0.9))
    Serie = scaler.fit_transform(Serie)  
    Serie_flat = Serie.ravel()
    energy = features.abs_energy(Serie_flat)
    length = features.length(Serie_flat)
    kurtosis = features.kurtosis(Serie_flat)
    variance = features.variance(Serie_flat)
    tree_examples.loc[i][['length','kurtosis','energy','variance']] = [length, kurtosis, energy, variance]
    
    
    logging.info("Empieza proceso genetico con FFNN")
    try:
        FFNN = FFNN_Function.main(steps_ahead,test_size,Serie,i)[0].mse
    except Exception as e:
        FFNN = -1
        logging.error("Exception occurred", exc_info=True)

    tree_examples.loc[i]['FFNN'] = FFNN
    tree_examples.to_csv(r'C:\Users\hug0er\Documents\Tesis\Meta-Table.csv')
    
    
    
    logging.info("Empieza proceso genetico con LSTM")
    try:
        LSTM = LSTM_Function.main(steps_ahead,test_size,Serie,i)[0].mse
    except Exception as e:
        LSTM = -1
        logging.error("Exception occurred", exc_info=True)

    tree_examples.loc[i]['LSTM'] = LSTM
    tree_examples.to_csv(r'C:\Users\hug0er\Documents\Tesis\Meta-Table.csv')
    
    
    
    logging.info("Empieza proceso de auto-ARIMA")
    try:
        ARIMA = ARIMA_Function.main(steps_ahead,test_size,Serie,i)[0].mse
    except Exception as e:
        ARIMA = -1
        logging.error("Exception occurred", exc_info=True)

    tree_examples.loc[i]['ARIMA'] = ARIMA
    tree_examples.to_csv(r'C:\Users\hug0er\Documents\Tesis\Meta-Table.csv')
    
tree_examples

```
