```python
def create_simple_cnn():  
  model = Sequential()
  model.add(Conv2D(16, kernel_size=(3, 3), input_shape=(656, 875, 1), activation='relu'))
  model.add(MaxPooling2D(pool_size=(2, 2)))

  model.add(Flatten())
  model.add(Dense(300))
  model.add(Dense(units=2, activation='softmax'))
  
  return model
```
